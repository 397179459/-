### 5.2 3DLine（3D折线/面积图）

```python
from pyecharts import Line3D

import math
_data = []
for t in range(0, 25000):
    _t = t / 1000
    x = (1 + 0.25 * math.cos(75 * _t)) * math.cos(_t)
    y = (1 + 0.25 * math.cos(75 * _t)) * math.sin(_t)
    z = _t + 2.0 * math.sin(75 * _t)
    _data.append([x, y, z])
# df = pd.DataFrame(np.random.rand(100,3))
# _data = df.values.tolist()
range_color = [
    '#313695', '#4575b4', '#74add1', '#abd9e9', '#e0f3f8', '#ffffbf',
    '#fee090', '#fdae61', '#f46d43', '#d73027', '#a50026']
line3d = Line3D("3D 折线图示例", width=790, height=400)
line3d.add(
    "",
    _data,
    is_visualmap=True,
    visual_range_color=range_color,
    visual_range=[0, 1],
    grid3d_rotate_sensitivity=5,
)
line3d
```

```python
## Q0.1 要求define一个函数，两个参数，第一是输入的数据源，第二个是绘制选取的字段名,默认为空
##  即全部绘制，绘制线性为 柱状图
df = pd.DataFrame(np.random.rand(10,4),columns = ['A1','A2','A3','A4'])
def PlotFunc(df,columns = []): 
    if columns == []:
        columns = df.columns
    if isinstance(columns,str):
        columns = [columns]
    temp =pd.Series(columns)
    temp1 = temp[~temp.isin(df.columns)]
    if len(temp1)!=0:
        print ("error",temp1.tolist())
        return 
    df1 = df[columns]
    
    line = Bar("折线图示例", width=790, height=400)
#     for i in df.columns:
#         line.add(i,df.index,df[i])
    [line.add(i,df1.index,df1[i]) for i in df1.columns]
    return line

PlotFunc(df,columns =['A1','A3','A2'])
```

### 5.12 Graph（关系图）

```python
from pyecharts import Graph

nodes = [{"name": "结点1", "symbolSize": 10},
         {"name": "结点2", "symbolSize": 20},
         {"name": "结点3", "symbolSize": 30},
         {"name": "结点4", "symbolSize": 40},
         {"name": "结点5", "symbolSize": 50},
         {"name": "结点6", "symbolSize": 40},
         {"name": "结点7", "symbolSize": 30},
         {"name": "结点8", "symbolSize": 20}]
links = []
for i in nodes:
    for j in nodes:
        links.append({"source": i.get('name'), "target": j.get('name')})
graph = Graph("关系图-力引导布局示例", width=790, height=400)
graph.add(
    "",
    nodes,
    links,
    is_label_show=True,
    graph_repulsion=8000,
    graph_layout="circular",
    label_text_color=None,
)
```

#### Q1 每个小时(以SHEET_START_TIME维度)每台Machine 的paticle走势 
#####  进阶要求 ： 能够在线进行对趋势数据 进行 线，柱，堆叠 的切换

```python
from pyecharts import *
import numpy as np
import pandas as pd

machine_list = data['MACHINENAME'].unique().tolist()
print(machine_list)

# machinename：指定看哪台设备的信息
def time_particle(df, machinename):
    line = Line('每小时每台Machine的paticle走势', width=990, height=400)
    df1 = df.groupby(['MACHINENAME', 'SHEET_START_TIME']).count()  
    # 根据 SHEET_START_TIME 重新根据半小时重采样 
    mac_resample = df1.loc[machinename].resample('0.5H')['LABEL'].sum()
    # 生成每半小时的 particle 数量
    var = mac_resample.tolist()
    # 生成横坐标，直接取 df1 的 index
    time = pd.Series(mac_resample.index).tolist()
    # 参数 is_more_utils 显示更多工具，默认只有下载/刷新/查看数据
    line.add(machinename, time, var, is_more_utils=True, is_label_show=True, is_smooth=False)    
    
    return line


time_particle(data, '2ATO09')
```

#### Q3 每个STEP 的paticle 在X方向上分布情况

```python
from pyecharts import *

def step_x_map(df):
    # 生成所有step作为横坐标
    x_axis = df['STEP'].unique().tolist()
    y_axis = []
    # 把每个STEP在X方向上的数据提取出来
    for x in x_axis:
        r = df[df['STEP'] == x]
        y_axis.append(r['X'].tolist())

    boxplot = Boxplot("每个STEP的paticle在X方向上的分布", width=990, height=1500/2)
    # 把数据转换成箱形图需要的格式
    _yaxis = boxplot.prepare_data(y_axis) 
    boxplot.add("boxplot", x_axis, _yaxis, is_more_utils=True)
    
    return boxplot

step_x_map(data)
```
