```python
# ### 习题 11
#     信用卡号合法性验证，校验的过程：
#     1、从卡号最后一位数字开始，逆向将奇数位(1、3、5等等)相加。
#     2、从卡号最后一位数字开始，逆向将偶数位数字，先乘以2（如果乘积为两位数，则将其减去9），再求和。
#     3、将奇数位总和加上偶数位总和，结果应该可以被10整除。
#     例如，卡号是：5432123456788881则，奇数位和=35
#     偶数位乘以2（有些要减去9）的结果：1 6 2 6 1 5 7 7，求和=35。
#     最后35+35=70 可以被10整除，认定校验通过。
#     请编写一个程序，从键盘输入卡号，然后判断是否校验通过。通过显示：“成功”，否则显示“失败”。
#     比如，用户输入：356827027232780
#     程序输出：成功
#     【参考测试用例】
#     356406010024817     成功
#     356827027232781     失败
# #### 同时判断以下三个信用卡号的合法性
# 358973017867744    
# 306406010024817    
# 358973017867754 
```

```python
# 构建一个test函数
def test(num):
    # 以下是函数主体
    x = 0
    y = 0
    l = list(list(str(num)))
    l.reverse()          # 反向之后更好地循环
    for i in range(len(l)):
        z = int(l[i])    # 转换成int型以便后续的运算
        if i%2 == 0:     # 判断奇偶性
            x = x + z
#             print(x)     # 监视奇数项的值
        else:
            z = z*2
            if z > 9:
                z = z-9
            y = y+z
#             print(y)     # 监视偶数项的值
    if (x+y)%10 == 0:
        print("总和为:",x+y)
        print("成功")
    else:
        print("总和为:",x+y)
        print("失败")
    return

# 键盘输入，调用test函数
a = eval(input("Please input your number: "))
test(a)

# 358973017867744  成功 
# 306406010024817  失败
# 358973017867754  失败
```

```python
# ### 名额分配问题
# 名额分配问题是 政治学中的一个数学问题，“按人口比例分配议员名额”的计算方法的问题，是数学在政治学中的一个应用。它以应用浅显的数学知识得出了深刻的政治结论，却一直未获根本解决，因此而著称于世。
#     把这个问题数学化，设美国一共有s个州，众议院一共设有h个议员席位。已知各州的人口，各州的议员席位按照各州人口比例划分，且各州至少保证有一个议员。
#     如何找出相应的一组整数a1，a2，…，as，使得 a1+a2+…+as=h，
#     让第i州取得a i（i=1, 2, 3,…, s）个议员名额，并且“尽可能地”满足美国宪法所规定的“按人口比例分配”的原则？这就是“名额分配问题”。从数学上说，稍加解释，小学生也可明白，但其求解却难倒了众多的政治家和数学家！
# ### 使用汉密尔顿法
#     1. 先让各州取得各州人口占比 乘以h的 qi 的整数部分 [qi]
#     2. 让 qi - [qi] 按照从小到大的顺序排列，将余下的名额逐渐分配给各相应的州，即小数最大部分的州优先分得余下的第一个名额，小数部分次大的州分得第二个名额，以此类推

# #### Ps:  Python 中 for 循环代码示例
#     a = ['A','B','C']
#     for i in a:
#         print (i)  ## i 是a中的每一个元素，即分别是 A,B,C
# ### 类比问题
# 某个大学的学院 有4个专业，学生会名额有20个，A系人数132，B系人数为157，C系人数为93，D系人数为76
```

```python
import math 
# 以下是题设的量，可以自由修改n的数值，字典中的元素
n = 20
d = {"A": 132, "B": 157, "C": 93, "D": 76}

# 以下是算法
l = list(d.values())
s = sum(l)        # 计算总人数
l2 = []
l3 = []

for i in l:
    i = (i/s)*n     # 计算出带小数的初值
    l1 = []
    l1 = math.modf(i)      # math.modf()函数把每一个浮点数拆分成[小数位，整数位]，便于下一步添加到两个新的列表中
    l2.append(l1[0])
    l3.append(int(l1[1]))   # int()可用可不用，为了好看

# 以上是循环建立两个新的列表，下面是找到小数部分大的，给对应的整数位+1
l4 = l3[:]             # 复制新的列表是为了实现对应的整数+1，因为每次循环要删除最大值
f = int(n - sum(l3))   # 计算出剩余几个名额，设定循环次数
for i in range(f):
    a = max(l2)
    b = l2.index(a)    # 找到小数列表最大值的下标
    c = l3[b]          # 在整数列表中找到对应的元素
    c += 1
    l4[b] = int(c)     # 用 + 1 之后的值替换整数列表中原来的元素
    del l2[b]
    del l3[b]          # 删除列表中最大的元素，继续找最大值

# 输出每个专业的学会生名额
print("每个专业的名额分别为:",l4)
```
